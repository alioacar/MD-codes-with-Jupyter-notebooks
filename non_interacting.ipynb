# A simple noninteracting particle MD

Don't forget to implement py3Dmol for visualization

```python
import numpy as np
import matplotlib.pyplot as plth
import matplotlib.pylab as plt
import py3Dmol
```

```python
params = {
    "natoms" : 100,
    "nsteps" : 1000,
    "box"    : ((0, 10), (0, 10), (0, 10)),
    "dt"     : 0.01
}
```

```python
def Add_Frame(positions,natoms,start):
    global trajectory
    
    if start == 0:
        trajectory = ''
    trajectory += str(natoms) + '\ntitle\n'
    for i in range(natoms):
        trajectory += ' '.join(['N',str(positions[i,0]),str(positions[i,1]),'0.0\n'])
```

```python
def Do_Step(positions,vels,dt):
    
    positions += vels*dt
```

```python
def Check_Boundary(positions,vels,box,dimension):
    
    for i in range(dimension):
        vels[((positions[:,i] <= box[i][0]) | (positions[:,i] >= box[i][1])),i] *= -1
```

```python
def run(natoms,nsteps,box,dt):
    
    dimension=len(box)
    positions=np.random.rand(natoms,dimension)
    for i in range(dimension):
        positions[:,i] = box[i][0] + (box[i][1]-box[i][0]) * positions[:,i]
    vels=np.random.rand(natoms,dimension)
    
    step=0
    while step <= nsteps:
        Do_Step(positions,vels,dt)
        Check_Boundary(positions,vels,box,dimension)
        if step%1000 == 0:  ### edit this to create a bigger trajectory
            Add_Frame(positions,natoms,step)        
        step +=1000
```

```python
run(**params)
```

```python
view = py3Dmol.view()
view.addModelsAsFrames(trajectory,'xyz')
#view.animate({'loop': 'forward', 'reps': 1}) #commented out because of HTML memory limit 
view.setStyle({'sphere':{'radius': 0.5}})
view.zoomTo()
```
